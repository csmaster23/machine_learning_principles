```python
import torch
import numpy as np
import torch.nn as nn
import matplotlib.pyplot as plt
```

```python
# 1. Introduction to Binary Cross Entropy Loss
# Binary Cross Entropy (BCE) is used to measure the performance of a binary
#    classification model whose output is a probability value between 0 and 1.

# Formula: BCE(y, y_pred) = - (y * log(y_pred) + (1 - y) * log(1 - y_pred))

# 2. Function to calculate BCE manually
def binary_cross_entropy(y_true, y_pred):
    """
    Manually calculates the binary cross entropy loss for a single true label and prediction.
    :param y_true: Ground truth label (1 or 0)
    :param y_pred: Predicted probability (must be between 0 and 1)
    :return: Binary Cross Entropy Loss
    """
    # Add epsilon to prevent log(0), which is undefined
    epsilon = 1e-15
    y_pred = np.clip(y_pred, epsilon, 1 - epsilon)

    # BCE loss formula: - (y_true * log(y_pred) + (1 - y_true) * log(1 - y_pred))
    loss = - (y_true * np.log(y_pred) + (1 - y_true) * np.log(1 - y_pred))
    return loss

# 3. Visualizing BCE Loss
# Generating probabilities (predictions) from 0 to 1
predictions = np.linspace(0.01, 0.99, 100)

# Calculate BCE loss for both class 1 (y_true = 1) and class 0 (y_true = 0)
loss_class_1 = [binary_cross_entropy(1, p) for p in predictions]
loss_class_0 = [binary_cross_entropy(0, p) for p in predictions]

# Plotting the BCE loss for different predictions
plt.figure(figsize=(10, 6))
plt.plot(predictions, loss_class_1, label="True Label = 1", color='red')
plt.plot(predictions, loss_class_0, label="True Label = 0", color='blue')
plt.title("Binary Cross Entropy Loss for True Labels 0 and 1")
plt.xlabel("Predicted Probability")
plt.ylabel("Binary Cross Entropy Loss")
plt.legend()
plt.grid(True)
plt.show()
```
